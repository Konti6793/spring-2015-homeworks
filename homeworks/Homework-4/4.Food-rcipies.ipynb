```python
#parse text file and turn into key value table (values are occurence of recpies, may be useful in later analysis)
text = []
words = {}
for line in open("./scirep-cuisines-detail/allr_recipes.txt"):
    text.append(line)
    line = line.split('\t')[1:]
    for word in line:
        if '\n' in word:
            word = word[:-1]
        if not word in words:
            words[word] = 0
        if word in words:
            words[word] = words[word]+1

words
```
