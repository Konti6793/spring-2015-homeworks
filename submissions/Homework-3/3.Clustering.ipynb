Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

```python
openDT = boston.OPEN_DT
closeDT = boston.CLOSED_DT
closeDT = closeDT.fillna(value='2015-01-01 00:00:01')

date = []
for dates in range(0, len(openDT)):
    open1 = openDT[dates].replace('-', '').replace(' ', '').replace(':', '')
    close1 = closeDT[dates].replace('-', '').replace(' ', '').replace(':', '')
    date.append((float(open1), float(close1)))

date = preproc.StandardScaler().fit_transform(date)


error = np.zeros(20)
error[0] = 0
for k in range (1,20):
    kmeans = KMeans(init = 'k-means++', n_clusters = k, n_init=10)
    kmeans.fit_predict(date)

    centroids = kmeans.cluster_centers_
    labels = kmeans.labels_
    error[k] = kmeans.inertia_
    
plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.ylabel('Error')
```

```python
#the slope of the curves steadies out around 4

kmeans = KMeans(init = 'k-means++', n_clusters = 4, n_init=10)
kmeans.fit_predict(date)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error[k] = kmeans.inertia_
    
print "The total error of the clustering is: \n", error
print '\nCluster labels'
print labels
print '\n Cluster Centroids'
print centroids

#plot
plt.scatter(date[:,0],date[:,1], c=labels)
```
