Given a social network of the characters in the marvel comicbook universe, and a list of all the marvel characters who are portrayed in movies, can a model be made that will predict if a given character will star in an upcoming movie.

IMPORTED LIBRARIES 

```python
import matplotlib as mpl
import numpy as np
import scipy as sp
import networkx as nx
import os
import seaborn as sns
import pandas as pd
```

PARSE DATA AND BUILD MASTERGRAPH

```python
#raw edge list
rawEL = open("edgeList.txt")
#raw node list
rawNL = open("nodeList.txt")

masterGraph = nx.Graph()

for line in rawNL:
    nodeNum = line.partition("\"")[0]
    nodeNum = int(nodeNum)
    nodeName = line.split("\"")[1]
    masterGraph.add_node(nodeNum, name = nodeName)

for line in rawEL:
    lineArr = line.split(" ")
    lineArr = map(int, lineArr)
    for i in range(1,len(lineArr)):
        masterGraph.add_edge(lineArr[0], lineArr[i])
        
```

```python
degree_sequence=sorted(nx.degree(masterGraph).values(),reverse=True) 
dmax=max(degree_sequence)
print dmax
```

All characters are connected to only comic books.  
Comic books are nodes past 6486.  All nodes conected to comic books will be linked, then the book node will be removed.

```python
charGraph = masterGraph
for book in range(6487,masterGraph.number_of_nodes()):
    #bookNode = charGraph.node[book]
    for character in charGraph.neighbors(book):
        for other in charGraph.neighbors(book):
            charGraph.add_edge(character,other)
    charGraph.remove_node(book)
charGraph.remove_node(19428)
```
